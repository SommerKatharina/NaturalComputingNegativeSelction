```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
EPS = [0.8, 0.5, 0.25, 0.15,  0.1,  0.05,  0.02, 0.01]
```

# Fitness Values Target = 1

```python
df = pd.read_csv('fitnesses.txt', header=None)
chunks = [df.iloc[:, i:i+20] for i in range(0, df.shape[1], 20)]

# Create a list to store new rows
new_rows = []

# Iterate over each chunk and create new rows
for chunk in chunks:
    new_row = pd.Series(chunk.values.flatten(), index=range(chunk.shape[1] * chunk.shape[0]))
    new_rows.append(new_row)

# Concatenate the new rows into a new DataFrame
df = pd.DataFrame(new_rows)

# Reset the index of the new DataFrame
df.reset_index(drop=True, inplace=True)

plt.hist(df.iloc[0], bins=15, alpha=0.7, label=EPS[0])
plt.hist(df.iloc[3], bins=15, alpha=0.7, label=EPS[3])
plt.hist(df.iloc[5], bins=15, alpha=0.7, label=EPS[5])
plt.hist(df.iloc[7], bins=15, label=EPS[7])
plt.xlabel('Fitness Score')
plt.ylabel('Density')
plt.title('Distribution of Fitness')
plt.legend()
plt.show()
```
